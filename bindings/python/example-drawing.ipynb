```python
import mata
import matplotlib.pyplot as plt
import networkx as nx
import scipy
import graphviz
import IPython
```

```python
lhs = mata.divisible_by(4)
rhs = mata.divisible_by(3)
result = mata.Nfa.concatenate(lhs, rhs)
```

```python
e_h = [
    (lambda aut, edge: edge.symb == 1, {'color': 'red'}),
    (lambda aut, edge: edge.src == edge.tgt, {'style': 'dashed', 'color': 'grey'})
]
n_h = [
    (lambda aut, state: state % 2 == 0, {'color': 'red'}),
    (lambda aut, state: state % 2 == 1, {'color': 'blue'}),
]
mata.plot(lhs, rhs, result, with_scc=True, node_highlight=n_h, edge_highlight=e_h)
```
